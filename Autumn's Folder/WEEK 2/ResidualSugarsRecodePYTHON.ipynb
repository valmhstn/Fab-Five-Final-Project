```python
# Recoding 'Residual Sugars' from Continuous to Categorical

## Classes of Still Red Wine
###  Up to 4 g/l---      Dry
###  4 g/l – 12 g/l---   Medium dry 
###  12 g/l – 45 g/l---  Medium (Medium sweet)
###  More than 45 g/l--- Sweet

## Recoding to:
### Dry -> 0
### Medium dry -> 1
### Medium -> 2
### Sweet -> 3
```

```python
# Import Packages

import pandas as pd
```

```python
# Load in Data

Wine = pd.read_csv ('/Users/autumnheyman/Desktop/WineQT.csv')
Wine.head()
```

```python
# Rename the Residual Sugars Column

Wine.rename(columns={'residual sugar' : 'residualsugar'}, inplace=True)
Wine.head()
```

```python
# Recode Residual Sugars

def residualsugar_recode (series): 
    if series <= 4.00:
        return 0
    if series <= 12.00: 
        return 1
    if series >= 12.01:
        return 2
    if series >= 45.01:
        return 3
    

Wine['residualsugarR'] = Wine['residualsugar'].apply(residualsugar_recode)
```
