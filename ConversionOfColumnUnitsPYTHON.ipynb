```python
# Columns where unit of measurement is necessary to update:

## Fixed acidity
## Volatile acidity
## Citric acid
## Chlorides
## Sulphates
```

```python
# Units of Measurement (Reference)

## 1 mg = 0.001 g
## 1 cm3 = 1 mL
## 1 dm3 = 1 L
## 1 g/dm3 = 1 g/L
## 1 mg/dm3 = 1 Part per Million
```

```python
# Load in Packages
import pandas as pd
import numpy as np
```

```python
# Load in Data
Wine = pd.read_csv ('/Users/autumnheyman/Desktop/WineQT.csv')
Wine.head()
```

```python
## Fixed Acidity

### Expressed in dataset as: tartaric acid, (g/dm3)
### Converting to: tartaric acid, (mg/L)
```

```python
### Rename Fixed Acidity Column

Wine.rename(columns={'fixed acidity' : 'fixedacidity'}, inplace=True)
```

```python
### Create a Fixed Acidity Recode Column

Wine['fixedacidityR'] = Wine['fixedacidity']*1000
```
